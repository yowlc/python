```python
import pandas as pd
import numpy as np

fd = pd.read_excel('學生成績單.xlsx')
fd.head()
```

```python
# Series 使用 map()

def grade(v):
    if v >= 80:
        return '優'
    elif v >= 70:
        return '甲'
    elif v >= 60:
        return '乙'
    else:
        return '丙'

fd['等級'] = fd['平均'].map(grade)
fd
```

```python
frame1 = fd[ ['國文', '英文', '數學', '地理', '歷史'] ]
frame1
```

```python
# DataFrame 使用 apply()

# 標準差
def fun3(s):
    return s.std()

frame1.apply(fun3)
```
