```python
import pandas as pd

df1 = pd.read_csv('world.csv')

df1[0:3]
# df1.head(3)
```

```python
# 洲名	國家	日期	總確診數	新增確診數	七天移動平均新增確診數	總死亡數	新增死亡數	...	七天移動平均疫苗新增接種劑數	每百人接種疫苗劑數	每百人接種疫苗人數	疫情控管指數	總人口數	中位數年紀	70歲以上人口比例	平均壽命	解除隔離數	解封指數

print(   df1[ ['洲名', '國家', '日期', '總確診數', '新增確診數', '總人口數', '解封指數'] ][:5]   )
print()

alldf = df1[ ['洲名', '國家', '日期', '總確診數', '新增確診數', '總人口數', '解封指數'] ]
print(alldf[:5])

print()

# 列出 只有 台灣 DataFrame
taiwan = alldf[ alldf['國家'] == '台灣' ]
print(taiwan[:5])
print()
print(taiwan[ taiwan['新增確診數'] >= 80000 ][:5])
print()
print(alldf[ (alldf['國家'] == '台灣') & (alldf['新增確診數'] >= 80000) ][:5])
```

```python
# 列出 前三筆
taiwan[ taiwan['新增確診數'] >= 50000 ][0:3]
```

```python
max1 = taiwan['新增確診數'].max()
print(max1)

taiwan[ taiwan['新增確診數'] == max1 ]
```

```python
taiwan[ taiwan['新增確診數'] == taiwan['新增確診數'].max() ]
```
