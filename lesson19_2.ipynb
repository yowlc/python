```python
import pandas as pd

df = pd.read_csv('學生成績.csv')
df
```

```python
df.index = df['學號']
display(df)

df1 = df.drop(columns='學號')
df1
```

```python
# 由到到小 排序
s1 = df1['平均'].rank(method='min', ascending=False)
df1['排名'] = s1
display(df1)

df2 = df1.sort_values(by='平均', ascending=False)
display(df2)

# 取 前 5 名
display(df2.head(n=5))
# display(df2[:5])
# display(df2.query('排名<=5'))

# 取 後 5 名
display(df2.tail(n=5))
# display(df2[-5:])
```

```python
# 輸出到 excel, 會出現 No module named 'openpyxl' 錯誤訊息
# pip install openpyxl 

df2.to_excel(excel_writer='學生成績單.xlsx', sheet_name='3年5班成績單')
```
