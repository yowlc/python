```python
import pandas as pd

df = pd.read_csv('各鄉鎮市區人口密度.csv')
df.head(n=5)
```

```python
df['區域別'].head(n=5)
```

```python
# Series.map() 修改完後, 可以存回原內容

def my_replace(v):
    return f'台灣-{v}'

df['區域別'].map(my_replace)
```

```python
def to_int(v):
    if v == '…':
        return 0
    return int(v)

df['年底人口數'] = df['年底人口數'].map(to_int)
df
```

```python
# 千位數 請加 , 號
def fun1(v):
    return f'{v:,}'

df['年底人口數'] = df['年底人口數'].map(fun1)
df
```

```python
def fun2(v):
    if v == '…':
        v = 0
    else:
        v = int(v)
    return f'{v:,}'

df['人口密度'] = df['人口密度'].map(fun2)
df
```
