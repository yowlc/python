```python
import pandas as pd
import numpy as np

s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
display(s1)
display(s2)
display(s3)
```

```python
# Series 合併

display( pd.concat([s1, s2, s3]) )
display( pd.concat([s1, s2, s3], axis=1) )
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2),
                   index=['a', 'b', 'c'],
                   columns=['one','two'])
display(df1)
df2 = pd.DataFrame(np.arange(4).reshape(2, 2),
                   index=['a', 'c'],
                   columns=['three','four'])
display(df2)
```

```python
# DataFrame 合併

pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=1, keys=['台北', '台南'])
```

```python
df1 = pd.DataFrame(np.random.randn(3, 4), columns=['a', 'b', 'c', 'd'])
display(df1)
df2 = pd.DataFrame(np.random.randn(2, 3), columns=['b', 'd', 'a'])
display(df2)
pd.concat([df1, df2])
```

```python
# ignore_index=True -> 重新索引

pd.concat([df1, df2], ignore_index=True)
```

```python
import yfinance as yf

data1 = yf.download("2330.TW", start="2023-09-15", end="2023-10-03")
tw2330 = data1[ ['Adj Close'] ]

tw2330.columns = ['台積電']
display(tw2330)

data2 = yf.download("2303.TW", start="2023-09-15", end="2023-10-03")
tw2303 = data2[ ['Adj Close'] ]

tw2303.columns = ['聯電']
display(tw2303)
```
