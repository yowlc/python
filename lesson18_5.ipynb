```python
# 各鄉鎮市區人口密度.csv
url = 'https://sheethub.com/data.gov.tw/%E5%90%84%E9%84%89%E9%8E%AE%E5%B8%82%E5%8D%80%E4%BA%BA%E5%8F%A3%E5%AF%86%E5%BA%A6?format=csv'

import pandas as pd

frame1 = pd.read_csv(url)
display(frame1.head(n=10))
```

```python
frame1.index = frame1['區域別']
frame1.head()
```

```python
frame2 = frame1[ ['年底人口數', '土地面積', '人口密度'] ]
frame2.head()
```

```python
errDf = frame2[frame2['年底人口數'] == '…']
display(errDf)

display(errDf.index)

errList = list(errDf.index)
display(errList)

frame3 = frame2.drop(index=errList)
```

```python
# 將 字串 轉成 數字
frame3['年底人口數'] = frame3['年底人口數'].astype(int)
frame3['土地面積'] = frame3['土地面積'].astype(float)
frame3['人口密度'] = frame3['人口密度'].astype(int)

frame3.info()
```
