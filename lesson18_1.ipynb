```python
import pandas as pd

df1 = pd.read_csv('world.csv')
df1.info()
```

```python
world_frame = df1[ ['洲名', '國家', '日期', '總確診數', '新增確診數', '總人口數', '解封指數'] ]
world_frame[:3]
```

```python
# 使用 [] subscript 方法 進行 搜尋

asia_frame = world_frame[ world_frame['洲名'] == '亞洲' ]
asia_frame[:5]
```

```python
world_frame[ (world_frame['國家'] == '台灣') & (world_frame['新增確診數'] == 0) ]
```

```python
# 使用 query 搜尋

query_frame = world_frame.query('國家=="西班牙" | 國家=="台灣"')

display(query_frame.head(n=3))
display(query_frame.tail(n=3))
```

```python
query1 = world_frame.query('國家=="台灣" and 新增確診數==0')
query1
```

```python
world_frame.query('日期=="2022-05-13" and 新增確診數<=50')
```

```python
world_frame.query('國家=="台灣" and 日期>="2022-07-01" & 日期<="2022-07-31"')
```

```python
taiwan = world_frame.query('國家=="台灣"')
tw_max = taiwan['新增確診數'].max()

# 要搜尋的內容在 變數, 要使用 @ 符號
taiwan.query('新增確診數==@tw_max')
```

```python
japan = world_frame.query('國家=="日本"')
jp_max = japan['新增確診數'].max()

# 要搜尋的內容在 變數, 要使用 @ 符號
japan.query('新增確診數==@jp_max')
```
