```python
# 多維資料 -> 樞紐 分析表

import numpy as np
import pandas as pd

# 多維 index -> stack 堆疊資料
s1 = pd.Series(np.random.randn(9),
          index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],
                 [1, 2, 3, 1, 2, 3, 1, 2, 3]])
s1
```

```python
s1.index
```

```python
s1.index.names = ['key1', 'key2']
s1
```

```python
display( s1.unstack(level='key1').fillna(0) )
display( s1.unstack(level='key2') )
```

```python
display( s1['a'][1] )
```

```python
frame = pd.DataFrame(np.arange(12).reshape((4,3)),
             index=[['a','a','b','b'], [1, 2, 1, 2]],
             columns=[['台北','台北','台中'], ['Green','Red','Green']])
frame.index.names = ["key1", "key2"]
frame.columns.names = ["縣市", "顏色"]
frame
```

```python
tmpdf1 = frame.stack(level=['縣市', '顏色'])
display(tmpdf1)

tmpdf2 = tmpdf1.unstack(level=['縣市', 'key1'])
display(tmpdf2)

# NaN 填入 0
display(tmpdf2.fillna(0))
tmpdf2.reset_index()
```

```python
# 統計

# 依據 index 加總
display(frame)

display(frame.groupby(level='key2').sum())

display(frame.groupby(level='key1').sum())
```

```python
display(frame)

frame.groupby(level='顏色', axis=1).sum()
```

```python
# 將 index 和 column 對調, 使用 大寫 T
display(frame)
display(frame.T)

frame.T.groupby(level='顏色').sum()
```

```python
frame = pd.DataFrame({'a':range(7), 'b':range(7, 0, -1),
                     'c':['one', 'one', 'one', 'two', 'two', 'two', 'two'],
                     'd':[0, 1, 2, 0, 1, 2, 3]})
frame
```

```python
# 把欄位 變成 索引  -> set_index()
# 把 索引 變成 欄位 -> reset_index()

frame2 = frame.set_index(['c', 'd'])
display(frame2)

display(frame2.reset_index())
```
