```python
#codespace 
#install -y
#:sudo apt-get update -y
#:sudo apt-get install -y fonts-wqy-zenhei
#:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
#  
#mac 
# font  = FontProperties(fname='/Library/Fonts/Arial Unicode.ttf', size=8)
# font1 = FontProperties(fname='/Library/Fonts/Arial Unicode.ttf', size=12)

# window
# font = FontProperties(fname=r'c:\windows\fonts\simsun.ttc', size=8)

# codespaces linux
font  = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc', size=8)
font1 = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc', size=12)

student1 = {'':78, '':80, '':92, '':75, '':85}
student1.keys()
student1.values()

student2 = {'':90, '':88, '':73, '':94, '':75}
student2.keys()
student2.values()

figure = plt.figure(figsize=[8, 5], facecolor="#cccccc")
axes = figure.add_subplot(1,1,1)
axes.plot(student1.keys(), student1.values(), 'r--+', label='1')
axes.plot(student2.keys(), student2.values(), 'b--.', label='2')

axes.set_xticklabels(student1.keys(), fontproperties=font)
axes.set_title('1', fontproperties=font)
axes.set_xlabel('', fontproperties=font)
axes.set_ylabel('', fontproperties=font)
axes.grid(axis='y')

axes.legend(prop=font)

student1_value_list = list(student1.values())
student2_value_list = list(student2.values())
xlocation = axes.get_xticks()
# ylocation = axes.get_yticks()
xtick_count = len(xlocation)
for i in range(xtick_count):
    x = xlocation[i]
    y1 = student1_value_list[i]
    y2 = student2_value_list[i]
    axes.annotate(str(y1), [x, y1])
    axes.annotate(str(y2), [x, y2])

plt.show()
```
