```python
import pandas as pd

df = pd.read_csv('學生成績.csv')
df
```

```python
# 搜尋 Series value

chinese = df['國文']
print(chinese[chinese >= 80])

math = df['數學']
print(math[math < 60])

df[ df['數學'] < 60 ]['數學']
```

```python
# 依據 dataframe 搜尋

print( df[ df['國文'] >= 80 ] )

print()

# 國文 和 數學 同時 不及格
print( df[ (df['國文'] < 60) & (df['數學'] < 60) ] )
```

```python
# 國文 英文 數學 地理 歷史 全部及格
df[ (df['國文'] >= 60) &
    (df['英文'] >= 60) &
    (df['數學'] >= 60) &
    (df['地理'] >= 60) &
    (df['歷史'] >= 60) ]
```

```python
df[ df['平均'] >= 80 ]
```

```python
# 國文 或 英文 不及格

d1 = df[ (df['國文'] < 60) | (df['英文'] < 60) ]
print(d1)

```
