```python
import numpy as np
import matplotlib.pyplot as plt

x_array = np.linspace(-3, 3, num=50)
print(x_array)
y1_array = x_array * 2 + 1
y2_array = x_array ** 2

figure = plt.figure(figsize=[8, 5])
axes = figure.add_subplot(1, 1, 1)
axes.plot(x_array, y1_array)
axes.plot(x_array, y2_array)

plt.show()
```

```python
import math
import matplotlib.pyplot as plt

t = np.arange(0, 2.5, 0.1)
print(t)
print(math.pi)

y1 = np.sin(math.pi * t)
y2 = np.sin(math.pi * t + math.pi / 2)
y3 = np.sin(math.pi * t - math.pi / 2)

figure = plt.figure(figsize=[8, 5])
axes = figure.add_subplot(1, 1, 1)
axes.plot(t, y1)
axes.plot(t, y2)
axes.plot(t, y3)

plt.show()
```
