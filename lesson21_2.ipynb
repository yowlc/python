```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/roberthsu2003/PythonForDataAnalysis/master/%E6%95%B8%E6%93%9A%E6%95%B4%E7%90%86_%E9%80%A3%E6%8E%A5_%E5%90%88%E4%BD%B5_%E9%87%8D%E5%A1%91/%E6%AD%B7%E5%B9%B4%E5%9C%8B%E5%85%A7%E4%B8%BB%E8%A6%81%E8%A7%80%E5%85%89%E9%81%8A%E6%86%A9%E6%93%9A%E9%BB%9E%E9%81%8A%E5%AE%A2%E4%BA%BA%E6%95%B8%E6%9C%88%E5%88%A5%E7%B5%B1%E8%A8%88.csv'

dataFrame = pd.read_csv(url,encoding='utf-8')
dataFrame.head()
```

```python
dataFrame.info()
```

```python
# 各年份, 每個縣市, 每月 的觀光人數

df1 = dataFrame.drop(columns=['類型', '觀光遊憩區', '細分'])
df1.head()
```

```python
df2 = df1.set_index(['年別', '縣市別'])
df2
```

```python
# 各年份, 每個縣市, 每月 的觀光人數

df3 = df2.groupby(level=['年別', '縣市別']).sum()
display(df3)
```

```python
df3.columns.name = '月份'
df4 = df3.stack()
display(df4)

df5 = df4.unstack(level=['縣市別', '月份'])
display(df5)

df5.to_excel('表單一.xlsx')
```

```python
# 欄位 -> 年別, 月份
df5 = df4.unstack(level=['年別', '月份'])
display(df5)
```
