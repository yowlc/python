```python
import requests

# 各鄉鎮市區人口密度.csv
url = 'https://sheethub.com/data.gov.tw/%E5%90%84%E9%84%89%E9%8E%AE%E5%B8%82%E5%8D%80%E4%BA%BA%E5%8F%A3%E5%AF%86%E5%BA%A6?format=csv'

#r = requests.request('get', url)
r = requests.get(url)

if r.ok:
    print('下載 OK')
```

```python
r.text
```

```python
import io
import csv
import pandas as pd

# 將 String 轉成 file
file = io.StringIO(r.text)

csv_reader = csv.reader(file)

# 轉成 list
data = list(csv_reader)
data[:5]
```

```python
df1 = pd.DataFrame(data)
display(df1[:5])

df1.columns = df1.iloc[0]
display(df1[:5])

df2 = df1.drop(index=0)
display(df2[:5])

display(df2.info())
display(df2['區域別'])
```

```python
frame1 = df2
frame1.index = frame1['區域別']
display(frame1[ ['年底人口數', '土地面積', '人口密度'] ].head(n=5))

# frame1.index.name = ''
display(frame1.head(n=5))
```

```python
# 兩種 方式 取得的新 DataFrame 都可以

frame2 = frame1.reindex(columns=['年底人口數', '土地面積', '人口密度'])
display(frame2.head(n=5))

frame3 = frame1[ ['年底人口數', '土地面積', '人口密度'] ]
display(frame3.head(n=5))
```

```python
display(frame2.info())

display(frame3.info())
```

```python
errDf = frame2[frame2['年底人口數'] == '…']
display(errDf)

display(errDf.index)

errList = list(errDf.index)
display(errList)
```

```python
#frame3 = frame2.drop(index=['東沙群島', '南沙群島'])
display(errList)
frame3 = frame2.drop(index=errList)

# 將 字串 轉成 數字
frame3['年底人口數'] = frame3['年底人口數'].astype(int)
frame3['土地面積'] = frame3['土地面積'].astype(float)
frame3['人口密度'] = frame3['人口密度'].astype(int)

frame3.info()
```
